# Importing 2024 Data File

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/2024 data_power demand delhi.csv')
```

```python
df
```

# Power Demand Trend

```python
power_demand=df.reset_index()['Maximum Demand Met (MW)']
```

```python
power_demand
```

```python
import matplotlib.pyplot as plt
plt.plot(power_demand, color='green', )
plt.xlabel('Days')
plt.ylabel('Power Demand (MW)')
plt.title('Power Demand in Delhi 2024')
plt.show()
```

# Weather Trends

## Maximum Temperature

```python
plt.plot(df['Maximum Temperature '])
plt.xlabel('Days')
plt.ylabel('Temperature (C)')
plt.title('Maximum Temperature in Delhi 2024')
plt.show()
```

## Minimum Temperature

```python
plt.plot(df['Minimum Temperature '])
plt.xlabel('Days')
plt.ylabel('Temperature (C)')
plt.title('Minimum Temperature in Delhi 2024')
plt.show()
```

## Maximum vs. Minimum Temperature

```python
plt.plot('Maximum Temperature ', data=df, color='red')
plt.plot('Minimum Temperature ', data=df, color='yellow')
plt.xlabel('Days')
plt.ylabel('Temperature (C)')
plt.title('Temperature in Delhi 2024')
```

## Humidity

```python
plt.plot(df['Humidity '])
plt.xlabel('Days')
plt.ylabel('Humidity (%)')
plt.title('Humidity in Delhi 2024')
plt.show()
```

## Precipitation

```python
plt.plot(df['Precipitation '])
plt.xlabel('Days')
plt.ylabel('Precipitation (mm)')
plt.title('Precipitation in Delhi 2024')
plt.show()
```

## Solar Radiation

```python
plt.plot(df['Solar Radiation '])
plt.xlabel('Days')
plt.ylabel('Solar Radiation (MJ/m^2)')
plt.title('Solar Radiation in Delhi 2024')
plt.show()
```

## Wind Speed

```python
plt.plot(df['Windspeed'])
plt.xlabel('Days')
plt.ylabel('Wind Speed (m/s)')
plt.title('Wind Speed in Delhi 2024')
plt.show()
```
